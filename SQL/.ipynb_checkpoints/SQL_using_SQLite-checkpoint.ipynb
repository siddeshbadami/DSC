/* Q10: Produce a list of facilities with a total revenue less than 1000.
The output of facility name and total revenue, sorted by revenue. Remember
that there's a different cost for guests and members! */

```python
df10=pd.read_sql_query('''select f.name as facility_name,
sum(CASE when b.memid!=0  then f.membercost*slots
 when b.memid=0 then f.guestcost*slots
 else 0 end) as revenue
from  Bookings b
inner join Facilities f
on b.facid=f.facid
group by facility_name
having revenue<1000
order by revenue''',conn)
```

```python
df10
```

Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order

```python
df11=pd.read_sql_query('''select surname,firstname
from Members
where recommendedby is not null
order by surname,firstname''',conn)
```

```python
df11
```

Q12: Find the facilities with their usage by member, but not guests */

```python
df12=pd.read_sql_query('''select distinct f.name,sum(slots)
from Bookings b
inner join Facilities f
on b.facid=f.facid
where b.memid!=0
group by f.name''',conn)
```

```python
df12
```

Q13: Find the facilities usage by month, but not guests

```python
df12=pd.read_sql_query('''select facid,extract(month from starttime) as month ,sum(slots)
from Bookings 
where memid!=0
group by month''',conn)
```
